```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels.formula.api as smf
import sklearn.metrics as sm
```

```python
btc = pd.read_csv('bitstampUSD_1-min_data_2012-01-01_to_2021-03-31.csv')
btc.sample(3)
```



***Convert timestamp***

```python
btc['Timestamp'] = pd.to_datetime(btc.Timestamp, unit='s')
btc.head()
```

```python
btc.tail()
```

**Find missing %**

```python
btc.isnull().sum().Open / btc.shape[0] * 100
```

***Remove NA value***

```python
btc.dropna(how='any', inplace=True)
```

```python
btc
```

**DatetimeIndex**

```python
btc.index = btc.Timestamp
btc.tail()
```

***See if one closed up or down each day. And create binary column 0 = loss, 1= gain to have something to predict.***

```python
btc['pos_neg'] = btc['Open']-btc['Close']
```

```python
btc['Up/Down'] = np.where(btc['pos_neg'] > 0, '0','1')
```

```python
btc.dtypes
```

```python
btc
```

```python
values = btc.values.astype('float64')
# Remove data that would not be known at prediction time ("high", "low", "close")
values = values[:, 0:4:]
```

```python
# Split data into test, train and validation with validation as most recent data
percent = int(values.shape[0]*0.01)
train = values[:percent*66, :]
val = values[percent*23:percent*90, :]
test = values[percent*90:percent*100, :]
```

```python
from sklearn import preprocessing
```

```python
# Split and scale data into train, val test sets
scaler = preprocessing.MinMaxScaler(feature_range=(0, 1))
train = scaler.fit_transform(train)
val = scaler.fit_transform(val)
test = scaler.fit_transform(test)
```

```python
X_train, y_train = train[:, :-1], train[:, -1]
X_val, y_val = val[:, :-1], val[:, -1]
X_test, y_test = test[:, :-1], test[:, -1]
```
